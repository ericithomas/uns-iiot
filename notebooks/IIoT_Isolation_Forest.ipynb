```python
import json, numpy as np, joblib
from sklearn.ensemble import IsolationForest
from google.colab import files

with open("baseline_data.json") as f:
    X = np.array(json.load(f))

print(f"Training on {X.shape[0]} windows, {X.shape[1]} features each")

model = IsolationForest(
    n_estimators=200,
    contamination=0.05,
    random_state=42,
)
model.fit(X)

baseline_scores = model.score_samples(X)
print(f"Baseline score range: {baseline_scores.min():.4f} to {baseline_scores.max():.4f}")
print(f"Baseline score mean: {baseline_scores.mean():.4f}")

joblib.dump(model, "isolation_forest.pkl")
files.download("isolation_forest.pkl")
```
